## Import and clean data

```python
import pandas as pd
```

```python
order_df=pd.read_excel('orders2.xlsx')
```

```python
order_df.head()
```

```python
order_df.info()
```

### Remove unwanted columns
Drop row id, country/region, ship mode, ship date 

```python
order_df=order_df.drop(['Row ID','Ship Date','Ship Mode','Country/Region'], axis=1)
order_df.head()
```

### Change column data type
Change the data type of Postal code column to object and Order Date to date time

```python
for col in order_df:
    if col =='Postal Code':
        order_df[col]=order_df[col].astype('object')
    elif col=='Order Date':
        order_df[col]=order_df[col].astype('datetime64')
order_df.info()
```

### Rename sales column

```python
order_df=order_df.rename({'Sales':'Revenue'}, axis=1)
order_df.head()
```

### Determine the most profitable product category

```python
order_df['Profit Margin']=order_df['Profit']/order_df['Revenue']
order_df.head()
```

```python
order_df_year=order_df.groupby([order_df['Order Date'].dt.year,'Category']).agg({'Profit':'sum'})
order_df_year.reset_index()
```

```python
order_df_pivot=pd.pivot_table(order_df,index=order_df['Order Date'].dt.year,columns='Category',values='Profit',aggfunc='sum')
order_df_pivot
```

```python
order_df_margin=order_df.groupby([order_df['Order Date'].dt.year,'Category']).agg({'Profit':'sum', 'Revenue':'sum'})
order_df_margin.reset_index()
```

```python
order_df_margin['PM']=order_df_margin['Profit']/order_df_margin['Revenue']*100
order_df_margin.columns=['Profit','Total Revenue','Profit_Margin%']
order_df_margin.reset_index()
```

```python
order_df_cummmargin=pd.pivot_table(order_df_margin, columns='Category',values='Profit_Margin%')
order_df_cummmargin
```

```python
order_df_margin=pd.pivot_table(order_df_margin, index='Order Date', columns='Category',values='Profit_Margin%')
order_df_margin
```

- No single category is the most profitable YOY although furniture is the least profitable of the 3
- Every category recorded a decline in profit margin for the year and percentage change from 2020 to 2021. Again, this may be associated with covid 19 lockdown
- Technology is the most profitable category cummulatively and edges out office suplies only slightly

### Determine profitability of the regional offices


```python
order_region=order_df.groupby([order_df['Order Date'].dt.year,'Region']).agg({'Profit':'sum'})
order_region.reset_index()
```

```python
order_region=pd.pivot_table(order_region, index='Order Date', columns='Region', values='Profit', aggfunc='sum')
order_region
```

```python
order_region_margin=order_df.groupby([order_df['Order Date'].dt.year, 'Region']).agg({'Profit':'sum','Revenue':'sum'})
order_region_margin.columns=['Profit','Revenue']
order_region_margin.reset_index()

#order_region_margin
```

```python
order_region_margin['Profit_Margin%']=(order_region_margin['Profit']/order_region_margin['Revenue'])*100
order_region_margin
```

```python
order_region_cumm_margin=pd.pivot_table(order_region_margin, columns='Region', values='Profit_Margin%')
order_region_cumm_margin
```

```python
order_region_margin=pd.pivot_table(order_region_margin, index='Order Date', columns='Region', values='Profit_Margin%')
order_region_margin
```

- All regions are profitable every year in review although the profit margin fluctuates. 
- Over the entire period, the most profitable region was the west for all years in review except 2020 
- The least profitable region over entire period is the central region and the most profitable is the west.

### Let us determine the most profitable category in each region

```python
order_cat_region=order_df.groupby(['Region','Category']).agg({'Profit':'mean'})
order_cat_region.reset_index()
```

```python
order_cat_region=pd.pivot_table(order_cat_region, index='Category', columns='Region', values='Profit', aggfunc='mean')
order_cat_region.reset_index()
```

- Furniture is the least profitable category across all regions except the South
- Technology is the most profitable category in all regions.

### Impact of discount on quantity ordered

```python
order_discount_qty=order_df.groupby(['Category', 'Discount Applied']).agg({'Quantity':'sum'})
order_discount_qty.reset_index()
```

```python
order_disc_region=pd.pivot_table(order_df,index='Category',columns=['Region', 'Discount Applied'], values='Quantity', aggfunc='sum')
order_disc_region
```

- Technology has the least sales volume across all regions while office supplies has the highest
- While increase in quantity when discount is applied can be seen, it does not automatically translate to profit and impact of discount the company's performance is better assesed from the profit metric.
- No definite trend is seen on the impact of discount on quantity across categories when assessed on a regional level region especially for furniture and technology categories.


## Impact of discount on profit

```python
order_discount=order_df.value_counts('Discount Applied',normalize=True )
order_discount
```

- The dataset has more discounted orders than regular orders

```python
order_discount=order_df.groupby(['Category','Discount Applied']).agg({'Profit':'sum','Revenue':'sum'})
order_discount.reset_index()
```

```python
order_discount=pd.pivot_table(order_discount, index='Category', columns='Discount Applied', values=['Profit','Revenue'], aggfunc='sum')
order_discount
```

- As expected, profit reduces for all categories when discount is applied but this particularly impacts furniture and office supplies category.
- However, sales/revenue increases for only 2 of the 3 categories while office supplies category saw a decline 
- Discount may need to be suspended or reduced for these categories.

```python
order_discount_reg=order_df.groupby(['Region','Category','Discount Applied']).agg({'Profit':'sum'})
order_discount_reg.reset_index()
```

```python
order_discount_reg1=pd.pivot_table(order_discount_reg, index='Region', columns=['Category','Discount Applied'], values='Profit', aggfunc='sum')
order_discount_reg1
```

```python
order_discount_reg2=pd.pivot_table(order_discount_reg, index='Region', columns='Discount Applied', values='Profit', aggfunc='sum')
order_discount_reg2
```

- As expected, applying discount wipes out all profits for all region except the west which also recorded a decline in profit but not a loss.
- Furniture category is impacted more when discounts are applied as all regions recorded losses for the category
